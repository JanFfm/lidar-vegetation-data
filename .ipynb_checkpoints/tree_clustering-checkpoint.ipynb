```python
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
from skimage.feature import peak_local_max
from skimage import img_as_float
from PIL import Image
import numpy
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline  
#setze gre der geplotteten bilder:
plt.rcParams['figure.figsize'] = [7, 7]

#lade bild als numpy.array
file = "ndom50_32282_5659_1_nw_2021.tif"
im = numpy.array(Image.open(file))
#scaliere in pixel-werte grer 0
m = - numpy.min(im)
im = im + m

""" #scaliere in range(0,255) und als integer-array:
m = numpy.max(im)
print(m)
im = ((im/m) *255).astype(numpy.int32)
"""
print(im)
plt.imshow(im)
plt.show()

#
im = numpy.where(im > 3, im, 0)
print(im)

plt.imshow(im)
plt.show()
```

```python
#im = img_as_float(im)
print(im)
image_max = ndi.maximum_filter(im, size=20, mode='constant')

# Comparison between image_max and im to find the coordinates of local maxima
coordinates = peak_local_max(im, min_distance=5)




# display results
fig, axes = plt.subplots(1, 3, figsize=(8, 3), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(im, cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(image_max, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Maximum filter')

ax[2].imshow(im, cmap=plt.cm.gray)
ax[2].autoscale(False)
ax[2].plot(coordinates[:, 1], coordinates[:, 0], 'r.')
ax[2].axis('off')
ax[2].set_title('Peak local max')

fig.tight_layout()

plt.show()

plt.imshow(im, cmap=plt.cm.gray)
plt.autoscale(False)
plt.plot(coordinates[:, 1], coordinates[:, 0], 'r.')
plt.axis('off')
plt.show()
```
