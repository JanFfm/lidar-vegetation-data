```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

im =cv2.imread("24007.png")
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
r,g,b = cv2.split(im)
kernel = np.ones((5,5),np.uint8)
i = 1
r = cv2.dilate(r,kernel,iterations = i)
g = cv2.dilate(g,kernel,iterations = i)
b = cv2.dilate(b,kernel,iterations = i)
im = cv2.dilate(im,kernel,iterations = i)

plt.imshow(r, cmap='gray')
plt.show()
plt.imshow(g, cmap='gray')
plt.show()
plt.imshow(b, cmap='gray')
plt.show()
plt.imshow(im)
plt.show()
```
